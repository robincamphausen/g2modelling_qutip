```python
# Define the positions in the autocorrelation where peaks should appear 
delayRange = 200
numPulses = int(2*np.floor(delayRange/12.5)+1)
print(f"numPulses = {numPulses}\n")

pulseInRange = np.floor(delayRange/12.5)*12.5
pulses = np.linspace(-pulseInRange,pulseInRange,numPulses)
print(f"pulses = {pulses}")
```

### Number of time elements

```python
# Resolution 100fs if *10000
# Resolution 1ps if *1000

numt = 2*delayRange*100+1 
numTau = 2*delayRange*100+1

print(f"{numt} time elements")
print(f"{numTau} tau elements")

# array of time (in ns)
tlist = np.linspace(-delayRange, delayRange, numt) 

# array of delays (in ns)
taulist = np.linspace(-delayRange, delayRange, numTau)
```

### Make photon flux pulsetrain

Each peak is an exponential decay with lifetime gamma


```python
# Repeat the list of time intervals as many times as there are laser pulses
tlist_matrix = np.repeat(tlist[:,None], len(pulses), axis=1)
tlist_matrix.shape

# Shift each one according to the pulse position
diffs = np.around(np.subtract(np.around(tlist_matrix,3), pulses),2)
diffs.shape

# Calculate negative exp for each one
exps = np.exp(-diffs/4.6)
exps.shape

# Put a cap to 1 (avoid probability higher than 1)
exps[exps > 1]=0

# Plot 4 of them
fig0 = plt.figure(figsize=(14,6))
ax01 = fig0.add_subplot(121)

for i in range(4):   
    _ = ax01.plot(tlist, exps[:,i])

# tlist_matrix[:5,:5]
# pulses[:5]
# diffs[:5,:5]
```

```python
import seaborn as sns
import pandas as pd 

def emit_photons(n_pulses=100000, lifetime=1):
    
    time_bins = np.arange(0,12.5*n_pulses,12.5)

    photons = np.random.exponential(scale=lifetime, size=n_pulses)
    timestamps = photons + time_bins
    
    return timestamps
    
    
def hbt_histogram_from_time(lifetime=1):

    timestamps = emit_photons(lifetime=lifetime)

    # Simulate splitting (randomly)
    beam_splitter = np.random.binomial(1, 0.5, size=timestamps.size).astype(bool)
    a = timestamps[beam_splitter]
    b = timestamps[~beam_splitter]

    A = np.ones([a.size,2])
    A[:,0] = a
    ch1 = pd.DataFrame(A, columns=['times', 'id'])

    B = np.zeros([b.size,2])
    B[:,0] = b
    ch2 = pd.DataFrame(B, columns=['times', 'id'])

    diff = pd.concat([ch1,ch2]).sort_values('times').diff().dropna()
    diff = diff[~(diff['id']==0)]
    differences = diff['times'].multiply(diff['id'])    
    
    return differences


def normalized_g2(lifetime):
    
    x_norm = np.linspace(-60,60,120)

    diffs = hbt_histogram_from_time(lifetime=lifetime)
    
    fig, ax = plt.subplots(1)
    sns.kdeplot(diffs, shade=True, ax=ax)

    # Extract y values
    y = ax.get_lines()[0].get_ydata()
    y = y/y.max()
    x = ax.get_lines()[0].get_xdata()

    y_norm = np.interp(x_norm, x, y)
        
    output = pd.Series(data=y_norm, index=x_norm, name=lifetime).to_frame()
    plt.close("all")
    
    return output


def find_best_tau(lifetimes, target_g20=0.5):
    
    simulated_g20 = np.zeros(len(lifetimes))
        
    for i, lt in enumerate(lifetimes):

        g2 = normalized_g2(lt)
        g2 = g2.reset_index()
        simulated_g20[i] = g2.iloc[(g2['index'] - 0).abs().argsort()[0], lt]
        
    print(f"{simulated_g20}")
    # Find tau resulting in a g2o
    idx = np.abs(simulated_g20 - target_g20).argmin()
    
    closest_g2 = simulated_g20[idx]
    closest_lifetime = lifetimes[idx]
    
    return closest_lifetime 
```

```python
import matplotlib.pyplot as plt
#linear fit

fig4 = plt.figure(figsize=(16,5))
ax40 = plt.subplot(121)
ax41 = plt.subplot(122)
#ax4.plot(decay_taulist,decayStats[2,:]/np.max(decayStats[2,:]),'o')

posDecays1000 = decay1000.T[12:18,:]
print(posDecays1000)
posDecays870 = decay870.T[12:18,:]
print(posDecays870)

posDecay_taulist = decay_taulist[12:18]
xp = np.linspace(posDecay_taulist[0],posDecay_taulist[-1],100)

#linear fit to log of the decay curve
fit1000 = np.polyfit(posDecay_taulist, np.log(posDecays1000), 1)
print(fit1000)
fit870 = np.polyfit(posDecay_taulist, np.log(posDecays870), 1)
print(fit870)

for q in range(fit1000[0,:].size):
    ax40.plot(xp,fit1000[0,q]*xp + fit1000[1,q])
    ax41.plot(xp,fit870[0,q]*xp + fit870[1,q])
    
    
#these are the decay lifetime parameters, and associated uncertainties:
print(np.mean(1/fit1000[0,:]))
print(np.std(1/fit1000[0,:]))

print(np.mean(1/fit870[0,:]))
print(np.std(1/fit870[0,:]))
```
